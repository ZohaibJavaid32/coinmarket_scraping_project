```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://coinmarketcap.com/"
request = requests.get(url)
```

```python
request
```

```python
soup = BeautifulSoup(request.content , 'html.parser')
```

```python
table = soup.find("table" ,class_='cmc-table')
table
```

```python
rows = table.find_all("tr")
```

```python
data = []
for row in rows[2:]:

    table_data = row.find_all('td')
    print(len(table_data))

    if len(table_data) < 6:
        continue

    rank = table_data[1].text
    name = table_data[2].text
    price = table_data[3].text
    hourly_change = table_data[4].text
    one_day_change = table_data[5].text
    weekly_change = table_data[6].text
    market_cap = table_data[7].text
    volume = table_data[8].text
    data.append((rank , name , price , hourly_change , one_day_change , weekly_change , market_cap ,volume))

#data
```
