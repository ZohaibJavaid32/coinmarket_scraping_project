```python
from bs4 import BeautifulSoup
import requests
# its used for html parsing.
# requests is used to access any API.
```

```python
url = 'https://realpython.github.io/fake-jobs/'
request = requests.get(url)
```

```python
# urlib = python library for parsing urls
```

```python
soup = BeautifulSoup(request.content,'html.parser')
#soup
```

```python
complete_title = soup.find('div', class_ = 'container mb-5')
print(complete_title.text.strip())
```

```python
title = soup.find('h1' ,class_ = 'title is-1')
print(title.text.strip())
```

```python
job_title = soup.find('h2' , class_ = 'title is-5').text.strip()
job_sub_title = soup.find('h3' , class_ = 'subtitle is-6 company').text.strip()
location = soup.find('p' , class_ = 'location').text.strip()
date = soup.find('p' , class_ = 'is-small has-text-grey').text.strip()
```

```python
print(job_title , job_sub_title , location ,date)
```

```python
# All divs

all_divs = soup.find_all('div', class_ = 'column is-half')
```

```python
def safe_text(tag):
    return tag.text.strip() if tag else None
```

```python
jobs = []
for job_html in all_divs:
    job = {
        "job_title":safe_text(job_html.find('h2', class_ = 'title is-5')),
        "company": safe_text(job_html.find('h3' , class_ = 'subtitle is-6 company')),
        "location": safe_text(job_html.find('p' , class_ = 'location')),
        "date": safe_text(job_html.find('p' , class_ = 'is-small has-text-grey')),
        "apply-link": job_html.find_all('a' , class_ = 'card-footer-item')[1]['href']
    }

    jobs.append(job)
    
    
```
